```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='ApproxHeight', data=df, hue='SitKCSE', s=100)
plt.title('Scatter plot of SitKSCE against Age and Height')
```

The above graph shows students of what height and age join any SitKCSE

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Course', y='Expense_Semester', data=df, palette='viridis')
plt.title('Course vs Expenses')
```

The above bar graph compares semster expenses of students from different courses
